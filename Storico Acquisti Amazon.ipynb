```python
# importiamo pandas e leggiamo i dati degli acquisti scaricati da Amazon

import pandas as pd

df = pd.read_csv(r"C:\Users\francesco.provenzano\Downloads\amazon_orders.csv")

df.head()
```

```python
df.shape # dimensioni dataframe
```

```python
# piccolo data cleaning: riempiamo i NaN

df = df.fillna(0)

df.head()
```

```python
# Devo controllare se il tipo dati per la colonna Total Owed (totale dovuto, i soldi che ho speso)
# è di tipo float per poter effettuare i calcoli. Se è string dovrò convertirlo

df.info()
```

```python
# è già di tipo float, perfetto.
```

```python
# Calcoliamo il totale speso.....

df["Total Owed"].sum()
```

```python
# Ho speso in totale 922,96€ in poco più di due anni. Sinceramente pensavo meno!

# Vediamo in media quanto ho speso per ordine:

df["Total Owed"].mean()
```

```python
# Il prezzo medio è stato di 19,63€. Considerando che compro soprattutto libri, mi sembra particolarmente azzeccato.
# Calcoliamo la mediana per neutralizzare gli outliers:

df["Total Owed"].median()
```

```python
# La mediana è 17,99 €, neanche a farlo apposta ci sarei riuscito volendolo.
# Poiché mediana > media, significa che ho fatto qualche particolarmente costoso, come ben ricordo.

df["Total Owed"].max()
# per curiosità vado a vedere quanto ho speso al max per un ordine
```

```python
# e al min
df["Total Owed"].min()
```

```python
# Guardiamo quanto ho speso di tasse!

df["Unit Price Tax"].sum()
```

```python
# "solo" 53€ di tasse, perché ad esempio i libri non ne hanno. Guardiamo al tax rate:

df["Unit Price Tax"].sum() / df["Total Owed"].sum()
```

```python
# Ora vogliamo rappresentare graficamente gli acquisti nel tempo. Per prima cosa convertiamo la colonna in formato datatime

df["Order Date"] = pd.to_datetime(df["Order Date"]).dt.tz_localize(None) # togliamo anche la timezone che non ci interessa
df.head()
```

```python
df.info() # verifichiamo che ora il dtype sia corretto
```

```python
# creiamo un primo grafico

df.sort_values(by=['Order Date'], ascending = True, inplace = True) # dal più vecchio al più recente, era il contrario

df.plot.bar(x='Order Date', y='Total Owed', rot=90, figsize=(20,10), title='Acquisti su Amazon')
```

```python
df_daily = df.groupby('Order Date').sum()['Total Owed'] # aggrego i giorni che hanno più acquisti

df_daily.head()
```

```python
df_daily.plot.bar(figsize=(20, 10), color='#61D199') # ricreo il grafico finale dove una barra = un singolo giorno aggregato
```
